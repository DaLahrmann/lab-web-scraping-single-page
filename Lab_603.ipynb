```python
import requests
import getpass
import pandas as pd

CLIENT_ID = 'fead378d2d834e79a84dff9c2c9e3776'
CLIENT_SECRET = getpass.getpass()
```

get access token for the spotify api

```python
AUTH_URL = 'https://accounts.spotify.com/api/token'

auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
```

request the spotify api

```python
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}

spotify=requests.get('https://api.spotify.com/v1/playlists/5pQVH0EJssqt4KDXPQVwv3/tracks',headers=headers)
```

get the list of tracks of the Ironhack Tracklist

```python
spotify1=spotify.json()
spoty2=pd.DataFrame(spotify1)
spoty3=pd.json_normalize(spoty2['items'])
spoty3.head(2)
```

Get `name`, `artists` and `track_id` of the Ironhack track list

```python
def artname(x):
    artlist=[]
    for i in x:
        artlist.append(i['name'])
    return artlist

artist=list(spoty3['track.artists'].apply(artname))   
tracks=list(spoty3['track.name'])
track_id=list(spoty3['track.id'])
spotylist=pd.DataFrame({'track_id': track_id,'track':tracks,'artist':artist})
display(spotylist)
spotylist.to_csv('Iron_Spotylist.csv',index=False)
```
